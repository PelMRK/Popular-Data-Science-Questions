## Getting the Data

The posts table has a lot of columns. We'll be focusing our attention on those that seem relevant towards our goal:

- `Id` - An identification number for the post


- `PostTypeId` - An identification number for the type of post
![types](data/types.png)


- `CreationDate` - The date and time of creation of the post


- `Score` - The post's score


- `ViewCount` - How many times the post was viewed


- `Tags` - What tags were used


- `AnswerCount` - How many answers the question got


- `FavoriteCount` - How many times the question was favored

Note that with exception of the tags column, the last few columns contain information about how popular the post is - the kind of information we're after.

Since we're only interested in recent posts, we'll limit our analysis to the posts of 2022. (At the time of writing it is end of summer 2022). 

To get the relevant data we run the following query.

```
SELECT Id, CreationDate,
       Score, ViewCount, Tags,
       AnswerCount, FavoriteCount
  FROM posts
 WHERE PostTypeId = 1 AND YEAR(CreationDate) = 2022;
```

## Exploring the Data

The result of query was stored in a file called `QueryResults.csv`. We can read in the data while immediately making sure `CreationDate` will be stored as a datetime object:

````python
### Preparing everything we need
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

questions = pd.read_csv("data/QueryResults.csv", parse_dates=["CreationDate"])
questions.info()
````

We see that only `FavoriteCount` has missing values. A missing value on this column probably means that the question was is not present in any users' favorite list, so we can replace the missing values with zero.

The types seem adequate for every column, however, after we fill in the missing values on `FavoriteCount`, there is no reason to store the values as floats.

Since the `object` dtype is a catch-all type, let's see what types the objects in `questions["Tags"]` are.

````python
questions["Tags"].copy().apply(lambda value: type(value)).unique()
````

We can see that every value in this column is a string.

## Cleaning the Data

We'll begin by filling in the missing values with `0`.

````python
questions.fillna(value={"FavoriteCount": 0}, inplace=True)
questions["FavoriteCount"] = questions["FavoriteCount"].astype(int)
questions.dtypes
````

Let's now modify `Tags` to make it easier to work with.

````python
questions["Tags"] = questions["Tags"].str.replace(r"^<|>$", "").str.split("><")
questions.sample(5)
````

## Most Used and Most Viewed

We now focus on determing the most popular tags. We'll do so by considering two different popularity proxies: for each tag we'll count how many times the tag was used, and how many times a question with that tag was viewed.

How many times each tag was used:

````python
tag_count = {}

for tags in questions["Tags"]:
    for tag in tags:
        if tag in tag_count:
            tag_count[tag] += 1
        else:
            tag_count[tag] = 1
                        
tag_count = pd.DataFrame.from_dict(tag_count, orient="index")
tag_count.rename(columns={0: "Count"}, inplace=True)
tag_count.head(10)
````

Let's now sort this dataframe by `Count` and visualize the top 20 results.

````python
most_used = tag_count.sort_values(by="Count").tail(20)
most_used
````

````python
most_used.plot(kind="barh", figsize=(16, 8))
````

Some tags are very broad and are unlikely to be useful, e.g.: `python`, `dataset`, `r`. Before we investigate the tags a little deeper, let's repeat the same process for views.

````python
tag_view_count = dict()

for index, row in questions.iterrows():
    for tag in row['Tags']:
        if tag in tag_view_count:
            tag_view_count[tag] += row['ViewCount']
        else:
            tag_view_count[tag] = row['ViewCount']
            
tag_view_count = pd.DataFrame.from_dict(tag_view_count, orient="index")
tag_view_count.rename(columns={0: "ViewCount"}, inplace=True)

most_viewed = tag_view_count.sort_values(by="ViewCount").tail(20)
````

````python
fig, axes = plt.subplots(nrows=1, ncols=2)
fig.set_size_inches((24, 10))
most_used.plot(kind="barh", ax=axes[0], subplots=True)
most_viewed.plot(kind="barh", ax=axes[1], subplots=True)
````

Looking at the results, we see that most top tags are present in both dataframes.

One way of trying to gauge how pairs of tags are related to each other, is to count how many times each pair appears together. Let's do this - the end result will be a dataframe that for each pair of tags, it tells us how many times they were used together.

````python
### Create a list of all tags
all_tags = list(tag_count.index)

### Create a dataframe where each row and column represents a tag
associations = pd.DataFrame(index=all_tags, columns=all_tags)
associations.fillna(0, inplace=True)

for tags in questions["Tags"]:
    associations.loc[tags, tags] += 1
    
associations.iloc[0:5,0:5]
````

````python
### Reduce the large dataframe to most used and viewed
relations_most_used = associations.loc[most_used.index, most_used.index]
relations_most_used
````

Let's create a heatmap to better get these relations.

````python
### Remove diagonal values
for i in range(relations_most_used.shape[0]):
    relations_most_used.iloc[i, i] = pd.np.NaN
    
plt.figure(figsize=(12, 8))
sns.heatmap(relations_most_used, cmap="Greens", annot=False)
````

The greener the cell, the more often these two tags occur together. For example, `deep-learning` is often used with `machine-learning`, `tensorflow` with `keras`, `pandas` with `python`.

## Just a Fad?

Before we officially make our recommendation, it would be nice to solidify our finding with additional proof. More specifically, one thing that comes to mind is "Is deep learning just a fad?". Ideally, the content we decide to create will be the most useful for as long as possible. Could interest in deep learning be slowing down?

The file `all_questions.csv` holds the result of the query below which fetches all of the questions ever asked on DSSE, their dates and tags:
```
SELECT Id, CreationDate, Tags
    FROM Posts
    WHERE PostTypeId = 1;
```
Here we will track the interest in deep learning across time. We will:

- Count how many deep learning questions are asked per time period


- The total amount of questions per time period


- How many deep learning questions there are relative to the total amount of questions per time period

````python
### Read in new data
all_q = pd.read_csv("data/all_questions.csv", parse_dates=["CreationDate"])
### Clean tag column as before
all_q["Tags"] = all_q["Tags"].str.replace("^<|>$", "").str.split("><")
````

Before deciding which questions should be classified as being deep learning questions, we should decide what tags are deep learning tags.

The definition of what constitutes a deep learning tag we'll use is: a tag that belongs to the list `["lstm", "cnn", "scikit-learn", "tensorflow", "keras", "neural-network", "deep-learning"]`.

This list was obtained by looking at all the tags in `most_used` and seeing which ones had any relation to deep learning.

````python
### Function that assigns `1` to deep learning questions and `0` otherwise
def class_deep_learning(tags):
    for tag in tags:
        if tag in ["lstm", "cnn", "scikit-learn", "tensorflow",
                   "keras", "neural-network", "deep-learning"]:
            return 1
    return 0


all_q["DeepLearning"] = all_q["Tags"].apply(class_deep_learning)
all_q.sample(5)
````

Looks good. The data-science-techonology landscape isn't something as dynamic to merit daily, weekly, or even monthly tracking. Let's track it quarterly.

````python
### Create a column that identifies the quarter in which a question was asked
def fetch_quarter(datetime):
    year = str(datetime.year)[-2:]
    quarter = str(((datetime.month-1) // 3) + 1)
    return "{y}Q{q}".format(y=year, q=quarter)


all_q["Quarter"] = all_q["CreationDate"].apply(fetch_quarter)
all_q.sample(5)
````

````python
### Group by quarter and calculate DL proportion

quarterly = all_q.groupby('Quarter').agg({"DeepLearning": ['sum', 'size']})
quarterly.columns = ['DLQuestions', 'TotalQuestions']
quarterly["DLProportion"] = quarterly["DLQuestions"] / quarterly["TotalQuestions"]
### The following is done to help with visualizations later.
quarterly.reset_index(inplace=True)
quarterly.sample(5)
````

````python
ax1 = quarterly.plot(x="Quarter", y="DLProportion", kind="line", 
                     linestyle="-", marker="o", color="orange",
                    figsize=(24, 12))

ax2 = quarterly.plot(x="Quarter", y="TotalQuestions", kind="bar",
                     ax=ax1, secondary_y=True, alpha=0.7, rot=45)

for idx, t in quarterly["TotalQuestions"].iteritems():
    ax2.text(idx, t, str(t), ha="center", va="bottom")

ax1.get_legend().remove()

handles1, labels1 = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(handles=handles1 + handles2,
           labels=labels1 + labels2,
           loc="upper left", prop={"size": 12})

for ax in (ax1, ax2):
    for where in ("top", "right"):
        ax.spines[where].set_visible(False)
        ax.tick_params(right=False, labelright=False)
````
